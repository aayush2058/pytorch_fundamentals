# Replicating non-linear activation functions

```python
# Creating a tensor

A = torch.arange(-10, 10, 1, dtype = torch.float32)
A.dtype
```

```python
A
```

```python
# A is a straight line

plt.plot(A);
```

## Relu

```python
torch.relu(A)
```

```python
plt.plot(torch.relu(A))
```

### Relu formula:  x = max(0, data)


```python
def custom_relu(data: torch.tensor) -> torch.tensor:
    our_function = torch.maximum(torch.tensor(0), data) # Everything in torch module requires tensor
    return our_function
    
```

```python
custom_relu(A)
```

```python
plt.plot(our_relu(A))
```

## Sigmoid

```python
torch.sigmoid(A)
```

```python
plt.plot(torch.sigmoid(A))
```

### Sigmoid formula: x = 1 / (1 + torch.exp(-data))

```python
def custom_sigmoid(data: torch.tensor) -> torch.tensor:
    our_function = 1 / (1 + torch.exp(-data))
    return our_function
```

```python
custom_sigmoid(A)
```

```python
plt.plot(custom_sigmoid(A))
```
